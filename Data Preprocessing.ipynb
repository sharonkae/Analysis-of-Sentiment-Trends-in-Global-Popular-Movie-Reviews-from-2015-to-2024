```python
reviews_df = pd.read_csv(reviews_path, encoding="latin1")
```

```python
reviews_df["review_rating"] = pd.to_numeric(reviews_df["review_rating"], errors="coerce")
reviews_df["imdb_id"] = reviews_df["imdb_id"].astype(str)
```

```python
reviews_df.rename(columns={
    "imdb_id": "id",
    "review title": "review_title",
    "review": "review_text"
}, inplace=True)
```

```python
reviews_df.columns
```

```python
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

reviews_df["clean_review"] = reviews_df["review_text"].apply(clean_text)
```
